```python
import easygui

class CGPA_CALC:
    
    name='Default'
    department = 'Default'
    Sem = 0
    CGPA = 0
    total_courses_completed = 0
    cumulative_cg = 0
    def __init__(self, name='Default', department='Default'):
        CGPA_CALC.name = name  ;   CGPA_CALC.department = department 
        
    def courses_and_cg(self,c):
        CGPA_CALC.Sem+=1

        list_of_courses, list_of_cg = [] , []
        avg = 0 ; total = 0
        for i in range(0,len(c),2):
            list_of_courses.append(c[i])
            list_of_cg.append(c[i+1])
            total += float(c[i+1])
        avg = total/len(list_of_courses)
        
        CGPA_CALC.total_courses_completed += len(list_of_courses)
        
        CGPA_CALC.cumulative_cg += total
        
        CGPA_CALC.CGPA = CGPA_CALC.cumulative_cg / CGPA_CALC.total_courses_completed
        
        CGPA_CALC.CGPA = "{:.2f}".format(CGPA_CALC.CGPA)

        a="Semester: "+str(CGPA_CALC.Sem)+'\nName: '+self.name+'\nDepartment: '+self.department   
        a=a+"\nCourses taken by "+self.name+": "
        
        for i in range(0,len(list_of_courses),1):
            if ( i == len(list_of_courses)-1 ):
                a=a+list_of_courses[i]+"."
            else:
                a=a+list_of_courses[i]+", "
        avg1="{:.2f}".format(avg)
        a=a+"\nGPA earned:"+avg1+"\nCGPA:"+str(CGPA_CALC.CGPA)
        easygui.msgbox(a, title="Result")
        
    def calculate(self,obj):
        index=0
        while 1:
            
            if index==0:
                msg = "Enter Your Course Codes and GPA"
                title = "CGPA Calculator"
                fieldNames = ['Name','Department',"Course 1",'Course 1 GPA',"Course 2",'Course 2 GPA',"Course 3",'Course 3 GPA',"Course 4",'Course 4 GPA',"Course 5",'Course 5 GPA']
                fieldValues = []  
                fieldValues = easygui.multenterbox(msg,title, fieldNames)
                while("" in fieldValues) : 
                    fieldValues.remove("")
                if fieldValues[0]=='STOP':
                    break
                obj = CGPA_CALC(fieldValues[0],fieldValues[1])
                obj.courses_and_cg(fieldValues[2:])
            else:
                msg = "Enter Your Course Codes and GPA"
                title = "CGPA Calculator"
                fieldNames = ["Type 'STOP' to quit,\n'CONTINUE' to Continue","Course 1",'Course 1 GPA',"Course 2",'Course 2 GPA',"Course 3",'Course 3 GPA',"Course 4",'Course 4 GPA',"Course 5",'Course 5 GPA']
                fieldValues = []  
                fieldValues = easygui.multenterbox(msg,title, fieldNames)
                while("" in fieldValues) : 
                    fieldValues.remove("")
                if fieldValues[0]=='STOP':
                    break
                obj.courses_and_cg(fieldValues[1:])
            index+=1
a=CGPA_CALC()
a.calculate(a)
```
